##Generador de variables aleatorias no uniformes##
*Método de la transformada inversa*

1. Dado $f(x)$, calculamos:$$\int_{-∞}^{x} f(ð)dð$$
2. Suponer $R \sim U(0,1)$ y resolver : $\quad F(x) = R$, para obtener $x = F^{-1}(R)$

Apliquemos esto con un ejemplo:

Sea:

$$f(x) = \begin{cases}
\frac{2(x-a)}{(c-a)(b-a)} & \text{si} & a\leq x \leq b \\
-\frac{2(x-c)}{(c-a)(c-b)} & \text{si} & b < x \leq c
\end{cases}$$

Entonces:
$$F(x) = \begin{cases}
\frac{(x-a)2}{(c-a)(b-a)} & \text{si} & a\leq x \leq b \\
1 - \frac{(x-c)2}{(c-a)(c-b)} & \text{si} & b < x \leq c
\end{cases}$$

Por lo que las inversas quedarían:
$$x = \sqrt{R(c-a)(b-a)} + a, \qquad \text{Para}\quad  a\leq x \leq b  $$

$$x = c - \sqrt{(1-R)(c-a)(c-b)}  \qquad \text{Para}\quad  b\leq x \leq c $$

```python
import random as rd
import matplotlib.pyplot as plt
import math

def exp1(a,b,c,n):
  x = []               # Generamos una lista vacía
  for i in range(n):   # Generamos valores aleatorios que tomará nuestra R para evaluarla en x
    R = rd.random()
    equis = math.sqrt(R*(c-a)*(b-a)) + a
    if a < equis < b:  # Específicamos el primer intervalo
      x.append(equis)  # Guardamos en nuestra lista los valores q tome x, según su función inversa correspondiente al primer intervalo
    else:
      equis = c-math.sqrt((1-R)*(c-a)*(c-b)) # Corresponde a nuestro segundo intervalo
      x.append(equis)  # Guardamos en nuestra lista los valores q tome x, según su función inversa correspondiente al segundo intervalo
  return x    # Returnamos todos los x de ambos intervalos

#sea a = 1, b = 3, c = 5
a = exp1(1,3,5,10000)  # Utilizando una n lo suficientemente grande, definimos una variable para tomar esa muestra de tamaño n
plt.hist(a, bins=50, color='cyan', edgecolor='black') # Generamos el histograma
plt.title('Distribución Triangular Generada')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.show()
```
