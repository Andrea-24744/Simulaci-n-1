```python
def generar_(a, c, m, X_o):
    n = 50
    X = [0] * n
    X[0] = X_o
    for i in range(1, n):
        X[i] = ((a * X[i - 1] + c) % m) / m
    return X

a = 4
c = 7
m = 22
X_o = 19
X_values = generar_(a, c, m, X_o)

# Imprimir en formato de tabla
print(f"{'n':<5}{'X[n]':<15}{'X[n]/m'}")
for i in range(len(X_values)):
    print(f"{i:<5}{X_values[i]:<15}{X_values[i]/m}")
```
