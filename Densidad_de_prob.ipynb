```python
import numpy as np
from tabulate import tabulate
import pandas as pd
import matplotlib.pyplot as plt

def generar(a, c, m, X_o):
    return (a * X_o + c) % m

a,c,m,X_o = 10001,821,10000,4
muestra = []
for i in range(10000):
    X_o = generar(a, c, m, X_o)
    muestra.append([i, X_o, X_o / m])

df = pd.DataFrame(muestra, columns=["n", "X_n", "X_n/m"])

desviacion = df["X_n/m"].std()

def estilo(valor):
    return 'color: green'

media = [np.mean([fila[2] for fila in muestra[:i+1]]) for i in range(len(muestra))]
varianza = [np.var([fila[2] for fila in muestra[:i+1]]) for i in range(len(muestra))]
numeros = sorted(df["X_n/m"])
prob_ac = np.cumsum(np.ones(len(numeros))) / len(numeros)


plt.hist([fila[2]for fila in muestra],bins=10)
plt.xlabel('n')
plt.ylabel('X_n')
plt.title('X_n/m')
plt.show()

#Graficar la media y la varianza al cuadrado
#plt.plot(range(len(media)), media, label='Media', color='blue')
#plt.plot(range(len(varianza)), np.square(varianza), label='Varianza al cuadrado', color='red')

#plt.xlabel('n')
#plt.ylabel('Valor')
#plt.title('Media y Varianza al Cuadrado')
#plt.legend()
#plt.show()

plt.plot(numeros, prob_ac, color='orange')
plt.xlabel('X_n/m')
plt.ylabel('Probabilidad acumulada')
plt.title('Funcin de Distribucin Acumulada (CDF)')
plt.grid(True)
plt.show()
```
