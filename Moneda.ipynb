```python
import matplotlib.pyplot as plt
import random

def volado(n):
    """Simula n volados y retorna la proporción de águilas."""
    aguila = sum(1 for _ in range(n) if random.random() < 0.5)  # 50% de águila
    return aguila / n  # Probabilidad de obtener águila

def experimento(m):
    """Realiza m experimentos y almacena la probabilidad obtenida en cada uno."""
    p = []
    for j in range(1, m + 1):  # Empezar desde 1 hasta m
        p.append(volado(j))  # Llamar correctamente a volado y guardar el resultado
    return p

# Parámetro de número de experimentos
m = 100  # Cambia esto si quieres más experimentos

# Ejecutar el experimento
p = experimento(m)

# Gráfico
plt.figure(figsize=(12, 5))
plt.plot(p, marker='o', linestyle='-')
plt.axhline(y=0.5, color='r', linestyle='--', label="Esperado (0.5)")
plt.title("Probabilidad generada de obtener águila")
plt.xlabel("Número de experimento")
plt.ylabel("P(Águila)")
plt.legend()
plt.grid(True)
plt.show()

```
