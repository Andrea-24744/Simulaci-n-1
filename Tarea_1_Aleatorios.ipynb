# GENERADOR CONGRUENCIAL
######Un generador congruencial es un algoritmo que produce una secuencia de números pseudoaleatorios. Es uno de los métodos más antiguos y conocidos para generar números pseudoaleatorios.
En esta actividad la relacion recursiva establecida es:
$$X_{n+1} = (aX_n + c)\text{mod}\hspace{.1cm}m$$
Conocida como generador lineal de congruencia.

Posterior a la generacion de los $X_n$ se observara el ciclo de vida de los valores obtenidos, el cual, no es más que la secuencia de números que produce antes de que se repita un patron cíclico, es decir, la cantidad de $X_n$ antes de que reaparezca cualquiera de las anteriores.

Finalmente para saber que efectivamente nuestro generador congruencial presenta su ciclo completo, se debe cumplir:

 1. $a$ debe ser entero no negativo, menor que $m$
 2. $(a - 1)\text{mod}\hspace{0.1cm}4 = 0$   si 4 es un factor de
 $m$
 3. $(a - 1)\text{mod}\hspace{0.1cm}b = 0$   si b es un factor primo de
 $m$

Explicado ya esto, programemos un ejercicio:

```python
#DOCUMENTADO#

from tabulate import tabulate
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Función de generador congruencial
def generar(a, c, m, X_o):
    return (a * X_o + c) % m

a,c,m,X_o = 81,89,100,5        # Definimos los valores de las constantes
muestra = []              # Lista vacía, para después almacenar los datos que generemos
muestra = [[i, X_o := generar(a, c, m, X_o), X_o / m] for i in range(108)]  # Se crean todos lo valores de x, segun el rango que establezcamos

df = pd.DataFrame(muestra, columns=["n", "X_n", "X_n/m"])     # Crea un formato de tabla de todos lo valores obtenidos por nuestro generador
print(tabulate(df, headers='keys', tablefmt='fancy_grid'))     # Imprime la tabla

#Función que nos arroje el ciclo de vida de nuestros datos
def ciclo_vida(lista):
    Aparición = {}            # Diccionario para registrar la primera aparición
    i = 0               # Inicializamos el índice

    while i < len(lista):         # Usamos while para recorrer todo valor
        _, X_n, _ = lista[i]       # De nuestra muestra (tabla), extraemos solamente la columna X_n

        if X_n in Aparición:       # Si ya apareció antes retornamos la longitud del ciclo y el índice

            return i - Aparición[X_n], i
        Aparición[X_n] = i        # Guardamos la posición donde apareció por primera vez
        i += 1              # Avanzamos al siguiente número

    return None            # Si no hay ciclo

V = ciclo_vida(muestra)
if V:                 # Veririca que el resultado tenga un valor distinto de NONE
    longitud_ciclo, indice = V          # Si hay un ciclo, "desempaqueta" la tupla es dos variables-salidas
    print(f"\nLongitud del ciclo: {longitud_ciclo}, comienza en el índice {indice - longitud_ciclo}")
else:
    print("\nNo se encontró un ciclo en la secuencia.")
```
