##Generador de variables aleatorias no uniformes##
*Método de la transformada inversa*

1. Dado $f(x)$, calculamos:$$\int_{-∞}^{x} f(ð)dð$$
2. Suponer $R \sim U(0,1)$ y resolver : $\quad F(x) = R$, para obtener $x = F^{-1}(R)$

Apliquemos esto con dos ejemplos:

Sea:

$$f(x) = \begin{cases}
\frac{2(x-a)}{(c-a)(b-a)} & \text{si} & a\leq x \leq b \\
-\frac{2(x-c)}{(c-a)(c-b)} & \text{si} & b < x \leq c
\end{cases}$$

Entonces:
$$F(x) = \begin{cases}
\frac{(x-a)2}{(c-a)(b-a)} & \text{si} & a\leq x \leq b \\
1 - \frac{(x-c)2}{(c-a)(c-b)} & \text{si} & b < x \leq c
\end{cases}$$

```python
import random as rd
import matplotlib.pyplot as plt
import math
#x = random.random()
def exp1(a,b,c,n):
  x = []
  for i in range(n):
    R = rd.random()
    equis = math.sqrt(R*(c-a)*(b-a)) + a
    if a < equis < b:
      x.append(equis)
    else:
      equis = c-math.sqrt((1-R)*(c-a)*(c-b))
      x.append(equis)
  return x

a = exp1(1,3,5,10000)
plt.hist(a, bins=50, color='cyan', edgecolor='black')
plt.title('Distribución Triangular Generada')
plt.xlabel('Valor')
plt.ylabel('Densidad')
print(np.mean(a))
print(np.var(a))
```
